```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#np.random.seed(42)
```

```python
#Function to generate random values for the function
def generate_values():
    X = np.random.uniform(low=0,high=1,size=300)
    V = np.random.uniform(low=-0.1,high=0.1,size=300)
    return X,V
```

```python
#Function to generate the given sin function
def generate_function():
    X,V = generate_values()
    D = [np.sin(20* x) + (3 * x) + v for x,v in zip(X,V)]
    return X,V,D
```

```python
#Function to plot the Input vs Output of the function
def plot(X,V,D):
    fig,ax = plt.subplots(figsize=(5,5))
    plt.scatter(X,D)
    plt.title('Plot of Input vs Output')
    plt.xlabel('X')
    plt.ylabel('D')
    plt.show()
    return
```

```python
#Function to plot the Input vs Output of the function and the predicted output by the network
def final_plot(X,V,D,Y2):
    fig,ax = plt.subplots(figsize=(8,8))
    plt.scatter(X,D,label = 'Actual')
    plt.scatter(X,Y2,label = 'Predicted',color='red')
    plt.title('Plot of Input vs Output')
    plt.xlabel('Input')
    plt.ylabel('Output')
    ax.legend(loc="best")
    plt.show()
    return
```

```python
#Function to plot the Mean Squared Error of the network at each epoch
def MSE_plot(epochs,errors):
    fig,ax = plt.subplots(figsize=(8,8))
    plt.plot(epochs,errors,label='Mean Squared Error')
    plt.title('Plot of Mean Squared Errors over the epochs')
    plt.xlabel('epochs')
    plt.ylabel('Mean Squared Error')
    ax.legend(loc="best")
    plt.show()
    return
```

```python
#Activation function at the output layer
def act_function(v):
    return v
```

```python
#activation function at the hidden layer - 1
def tanh_function(v,iteration):
    if iteration == 'forward':
        return np.tanh(v)
    return (1 - (np.tanh(v)) ** 2)
```

```python
#Function to generate weights and biases for the network
def generate_weights(n):
    w1 = np.random.normal(1,1,n).reshape(n,1)
    b = np.random.normal(0,1,n).reshape(n,1)
    w2 = np.random.normal(1,1,n).reshape(1,n)
    d = np.random.normal(0,1,1).reshape(1,1)
    W1 = np.concatenate((b,w1),axis = 1)
    W2 = np.concatenate((d,w2),axis = 1)
    return W1,W2
```

```python
#Function to implement forward and backward propogation and find the desired set of weights of the network
def forward_backward_prop(eta,X,D,W1,W2):
    errors = []
    epoch = 0
    epochs = []
    while epoch < 5000:
        Y2 = []
        for i in range(len(X)):
            #first layer
            unit = np.ones((1,1))
            x = X[i].reshape(1,1)
            v1 = np.matmul(W1,np.vstack((unit,x)))
            y1 = tanh_function(v1,'forward')
            #second_layer
            v2 = np.matmul(W2,np.vstack((unit,y1)))
            y2 = act_function(v2)
            Y2.append(y2)
            #backpropogation
            #E = (D[i] - y2) ** 2
            delta_2 = (D[i] - y2) * 1
            delta_1 = np.matmul(W2.transpose(),delta_2)[1:,:] * tanh_function(v1,'backward')
            dE_dW1 = np.matmul(-delta_1,np.vstack((unit,x.reshape(1,1))).transpose())
            dE_dW2 = np.matmul(-delta_2,np.vstack((unit,y1)).transpose())
            #update weights
            W1 = W1 + (eta * np.matmul(delta_1,np.vstack((unit,x.reshape(1,1))).transpose()))
            #print(delta_2.shape,W2.shape,y1.shape)
            W2 = W2 + (eta * np.matmul(delta_2,np.vstack((unit,y1)).transpose()))
        #calculate error    
        error = 0
        for i in range(len(D)):
            error+= (D[i] - Y2[i]) ** 2
        mse = error/len(D)
        errors.append(mse[0,0])
        #checking for iteration
        if epoch > 0:
            if mse <= 0.001:
                break
            if errors[epoch] > errors[epoch-1]:
                eta = eta * 0.9
        epochs.append(epoch)
        epoch+= 1
    return errors,epochs,W1,W2
```

```python
#Function implementing forward propogation to get the output for the input
def predict_function(X,W1,W2):
    Y2 = []
    for i in range(len(X)):
        unit = np.ones((1,1))
        x = X[i].reshape(1,1)
        v1 = np.matmul(W1,np.vstack((unit,x)))
        y1 = tanh_function(v1,'forward')
        #second_layer
        v2 = np.matmul(W2,np.vstack((unit,y1)))
        y2 = act_function(v2)
        Y2.append(y2)
    return Y2
```

```python
X,V,D = generate_function()
_ = plot(X,V,D)
W1,W2 = generate_weights(24)
print('Shape of W1: {} \nShape of W2: {}'.format(W1.shape,W2.shape))
```

```python
#Learning rate - 0.01 and Threshold is 0.001
eta = 0.01
errors,epochs,W1,W2 = forward_backward_prop(eta,X,D,W1,W2)
_ = MSE_plot(epochs,errors)
Y2 = predict_function(X,W1,W2)
```
